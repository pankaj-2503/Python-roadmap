First reading the pdf file and adding all text into varaible text using fitz library

```python
import fitz

def extract_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

pdf_path = "test3.pdf"
text = extract_text(pdf_path)
print(text)
```

Saved extracted data into excel using dataframe to excel 

```python
import pandas as pd
def save_text_to_excel(text,output_excel):
    lines=text.split("\n")
    df=pd.DataFrame(lines,columns=["text"])
    df.to_excel(output_excel,index=False)
    print(f"Data saved to {output_excel}")
text=extract_text(pdf_path)
output_excel="output.xlsx"
save_text_to_excel(text,output_excel)
```

```python
def text_to_dataframe(text):
    lines = text.split("\n")
    data = []
    current_key = None
    current_value = None

    for line in lines:
        # Skip separator lines like "-------" or empty lines
        if line.strip() == "" or set(line.strip()) == {"-"}:
            continue

        # Check if the line contains a colon
        if ":" in line:
            # If there's an ongoing key-value pair, save it before processing the new line
            if current_key is not None:
                data.append([current_key, current_value if current_value else None])
                current_key, current_value = None, None

            # Split the line into key and value
            parts = line.split(":", 1)
            current_key = parts[0].strip()
            current_value = parts[1].strip() if len(parts) > 1 else None
        else:
            # If the line does not contain a colon, treat it as a continuation of the current value
            if current_value is not None:
                current_value += f" {line.strip()}"
            else:
                current_value = line.strip()

    # Add the last key-value pair if it exists
    if current_key is not None:
        data.append([current_key, current_value if current_value else None])

    # Convert the list of rows into a DataFrame
    df = pd.DataFrame(data, columns=["Field", "Value"])
    return df
text=extract_text(pdf_path)
df=text_to_dataframe(text)
df.tail(20)
```

```python
output_excel = "cleaned_data.xlsx"
df.to_excel(output_excel, index=False)
print(f"Data saved to {output_excel}")
```
