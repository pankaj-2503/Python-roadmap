```python
data=data[["Product_title","Product_price","Currency","Product_num_ratings","Is_best_seller"]]
```

```python
data["Is_best_seller"]=data["Is_best_seller"].apply(lambda x: 1 if x==True else 0)
data
```

```python
data["Product_price"]=data["Product_price"].str.replace("$","")
data
```

```python
data["Product_price"]=data["Product_price"].astype(float)
data
```

```python
data["Product_price"]=data["Product_price"].apply( lambda x: round(x) )
data
```
