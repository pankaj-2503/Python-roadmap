```python
# filter out only int and float valued columns
numeric_cols=df.select_dtypes(include=["int","float"]).columns
corr_matrix = df[numeric_cols].corr()
print(corr_matrix)
```

```python
plt.figure(figsize=[10,7])
sns.heatmap(corr_matrix,annot=True,cmap='plasma',linewidths=5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
position_count=df["Position"].value_counts()
position_count
# index is pos position_count
```

```python
myexplode=[0.13,0,0,0,0]
plt.pie(position_count,labels=position_count.index,autopct="%1.1f%%",startangle=90,shadow=True,explode=myexplode)
plt.title("Position Distribution")
plt.show()
```

```python
# average attack for each country
average_attack_by_country=df.groupby("Country")["Attack"].mean()
average_attack_by_country.sort_values(ascending=False)
```

```python
plt.figure(figsize=[8,6])
average_attack_by_country.sort_values(ascending=False).plot(kind='bar',color='red',linewidth=8)
plt.title("Average Attack by Each Country")
plt.xlabel("Country")
plt.ylabel("Attack")
plt.grid(color='grey',linewidth=0.2,linestyle='--')
plt.show()
```

```python
average_serve_by_age=df.groupby("Age")["Serve"].mean()
average_serve_by_age.sort_values(ascending=False)
```

```python
# If we have to correct the columns like if it has space then 
df.rename(columns={"Receive":"Receive"},inplace=True)
```

```python
plt.scatter(df["Block"],df["Receive"],color="hotpink")
plt.title("Scatter plot: Block vs Receive")
plt.xlabel("Blocks")
plt.ylabel("Received")
plt.grid(linestyle="--",color="black")
plt.show()
```

```python
plt.hist(df["Age"],bins=20,color="skyblue",edgecolor="black")
plt.grid()
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
average_attack_by_position=df.groupby("Position")["Attack"].mean()
average_attack_by_position.sort_values(ascending=False).plot(kind='bar',color="green")
plt.title("Average Attack By Position")
plt.xlabel("Position")
plt.ylabel("Average Attack")
plt.grid(color="grey",linestyle="--")
plt.show()
```

```python
serve_trend_by_age=df.groupby("Age")["Serve"].mean()

serve_trend_by_age.plot(kind="line",marker='o',linestyle="-",color="orange")
plt.grid(linestyle='--')
plt.title("Serve Trend Over Age Group")
plt.xlabel("Age")
plt.ylabel("Average Serve")
plt.show()
```

```python
total_attack_block_by_country=df.groupby("Country")[["Attack","Block"]].sum()
total_attack_block_by_country
```

```python
total_attack_block_by_country.sort_values(ascending=False,by="Attack").plot(kind="bar",stacked=True,colormap="viridis")
plt.title("Total Attack and Block by Country")
plt.xlabel("Country")
plt.ylabel("Total Value")
plt.grid()
plt.show()
```
